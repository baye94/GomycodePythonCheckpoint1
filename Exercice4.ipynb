```python
nomSaisi = input("Saisir un nom");
valeurSaisi = int(input(" veuillez saisir une valeur"));
print(len(nomSaisi))
if(valeurSaisi >=0 | valeurSaisi <= len(nomSaisi)):
    resultat = list(nomSaisi.strip())
    print("le caractere supprimer est: "+ resultat.pop(valeurSaisi))
    nomSaisi = "" .join(resultat);
    print("le resultat est : "+nomSaisi);
else:
    print("la valeur que vous avez saisie est superieur aux nombre de caractere de la chaine : "+nomSaisi )
```
