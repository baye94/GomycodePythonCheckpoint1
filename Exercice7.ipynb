```python
from math import *
valeurc= int(50)
valeurh=int(30)
tableaux= list()
def calcule_formule(chaineSaisie):
    print( " Les valeurs saisies sont :"+ str(chaineSaisie))
    nouveauxchaine = chaineSaisie.rsplit(",")
    r =range(0,len(nouveauxchaine))
   
    for i in r:
        print(" pour la valeur " +str(nouveauxchaine[i])+  " -> "+ str(round(sqrt((2*valeurc*int(nouveauxchaine[i]))/valeurh))) )
        tableaux.append(str(round( sqrt ((2*valeurc*int(nouveauxchaine[i]))/valeurh))))
    
    chaineFinale = ','.join(tableaux)
    print( " Les valeurs apres les calcules sont :"+ str(chaineFinale))
    
x = input('donnez une serie de valeur separer par des virgule')
calcule_formule(x)
```
